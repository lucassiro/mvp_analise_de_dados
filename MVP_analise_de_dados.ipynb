# 1. Definição do problema

O dataset que será usado nesse projeto é o [**Student Performance Factors**](https://www.kaggle.com/datasets/lainguyn123/student-performance-factors), que possui licença [**CC0: Public Domain**](https://creativecommons.org/publicdomain/zero/1.0/).  
  
Este conjunto de dados fornece uma visão geral abrangente de vários fatores que afetam o desempenho dos alunos em exames. Inclui informações sobre hábitos de estudo, frequência, envolvimento dos pais e outros aspectos que influenciam o sucesso acadêmico.

Esse dataset é composto dos seguintes atributos:

| Atributo | Descrição |
|--|--|
| Hours_Studied | Número de horas gastas estudando por semana. |
| Attendance | Porcentagem de aulas assistidas. |
| Parental_Involvement | Nível de envolvimento dos pais na educação do aluno (Low, Medium, High). |
| Access_to_Resources | Disponibilidade de recursos educacionais (Low, Medium, High). |
| Extracurricular_Activities | Participação em atividades extracurriculares (Yes, No). |
| Sleep_Hours | Número médio de horas de sono por noite. |
| Previous_Scores | Pontuações de exames anteriores. |
| Motivation_Level | Nível de motivação do aluno (Low, Medium, High). |
| Internet_Access | Disponibilidade de acesso à internet (Yes, No). |
| Tutoring_Sessions | Número de sessões de tutoria assistidas por mês. |
| Family_Income | Nível de renda familiar (Low, Medium, High). |
| Teacher_Quality | Qualidade dos professores (Low, Medium, High). |
| School_Type | Tipo de escola frequentada (Public, Private). |
| Peer_Influence | Influência dos colegas no desempenho acadêmico (Positive, Neutral, Negative). |
| Physical_Activity | Número médio de horas de atividade física por semana. |
| Learning_Disabilities | Presença de dificuldades de aprendizagem (Yes, No). |
| Parental_Education_Level | Maior nível de educação dos pais (High School, College, Postgraduate). |
| Distance_from_Home | Distância de casa até a escola (Near, Moderate, Far). |
| Gender | Gênero do aluno (Male, Female). |
| Exam_Score | Pontuação do exame final. |

### 3.2 Seleção das colunas

Das 20 colunas:
* 7 possuem dados numéricos (discretos e contínuos)
* 13 possuem dados categóricos

Aqui irei dividir as colunas do dataset em duas partes, uma com as variáveis categóricas e outra com as variáveis numéricas e plotarei tipos de gráficos diferentes para cada uma delas.

```python
numerical = [
    "Hours_Studied",
    "Attendance",
    "Sleep_Hours",
    "Previous_Scores",
    "Tutoring_Sessions",
    "Physical_Activity",
    "Exam_Score",
]

categorical = [
    "Motivation_Level",
    "Parental_Involvement",
    "Access_to_Resources",
    "Extracurricular_Activities",
    "Internet_Access",
    "Family_Income",
    "Teacher_Quality",
    "School_Type",
    "Peer_Influence",
    "Learning_Disabilities",
    "Parental_Education_Level",
    "Distance_from_Home",
    "Gender",
]
```

### 3.3. Visualizações Unimodais
Aqui temos um histograma para cada atributo numérico discreto e contínuo do dataset.

```python
# Histograma
df[numerical].hist(figsize=(15, 10))
plt.show()
```

Aqui temos gráficos de densidade, que são parecidos com o histograma e nos proporcionam uma visualização um pouco diferente.

```python
# Density Plot
df[numerical].plot(kind="density", subplots=True, layout=(3, 3), sharex=False, figsize=(15, 10))
plt.show()
```

E o nosso último gráfico unimodal é o boxplot.
Na linha do centro temos a mediana, a linha de baixo é o primeiro quartil e a linha de cima o terceiro quartil.

```python
# Boxplot
df[numerical].plot(kind="box", subplots=True, layout=(3, 3), sharex=False, sharey=False, figsize=(15, 10))
plt.show()
```

### 3.4. Visualizações Multimodais

A primeira visualização aqui será um mapa de calor das correlações de pearson para as variáveis numéricas.  
Podemos ver que Attendance (porcentagem de aulas assistidas), e Hours_Studied (número de horas gastas estudando por semana) são as variáveis numéricas com a maior correlação com o Exam_Score (nota do exame).  
Isso nos mostra o quanto é importante estudar bastante e assistir todas as aulas.

```python
# Matriz de Correlação com Matplotlib Seaborn
sns.heatmap(df[numerical].corr(), annot=True, cmap="Blues")
plt.show()
```

No gráfico de dispersão podemos ver o relacionamento entre duas das variáveis numéricas.

```python
# Scatter Plot com Seaborn
sns.pairplot(df[numerical])
plt.show()
```

Aqui temos gráficos de barras para as variáveis categóricas, nos mostrando a distribuição entre as classes, e a partir desses gráficos podemos fazer algumas observações interessantes.

* A maioria dos alunos possuem acesso a internet.
* A maioria dos alunos estuda em escola pública.
* A maioria dos alunos vem de familias com renda média e baixa.
* Uma pequena parcela possui problemas de aprendizado.

```python
figsize = (15, 12)
rows = 4
columns = 4

fig = plt.figure(figsize=figsize)
for n, column in enumerate(categorical):
    temp_df = df[column].value_counts()

    fig.add_subplot(rows, columns, n + 1)
    plt.barh(temp_df.index, temp_df.values)
    plt.subplots_adjust(wspace=0.5, hspace=0.6)
    plt.title(column)
```

Podemos também visualizar as médias (mean), desvio padrão (std) e contagem (count) dos dados.

Aqui temos o nível de educação dos pais, podemos ver que é um fator que não influenciou nas notas, visto que a média das notas do exame ficou entre 66,89 e 67,97.

```python
df[["Parental_Education_Level", "Exam_Score"]].groupby("Parental_Education_Level").agg(["mean", "std", "count"])
```

Também podemos ver que a nota média dos alunos de escolas públicas ficaram quase iguais as notas dos alunos de escolas particulares.

```python
df[["School_Type", "Exam_Score"]].groupby("School_Type").agg(["mean", "std", "count"])
```

Gênero é outro fator que não influenciou nas notas, pois a nota média do gênero masculino (male) e feminino (female) foi praticamente a mesma. Podemos notar uma diferença ligeira no desvio padrão, onde as notas do gênero feminino variaram mais.

```python
df[["Gender", "Exam_Score"]].groupby("Gender").agg(["mean", "std", "count"])
```

Também não houve uma grande diferença nas médias dos alunos com e sem problemas de aprendizagem.

```python
df[["Learning_Disabilities", "Exam_Score"]].groupby("Learning_Disabilities").agg(["mean", "std", "count"])
```
